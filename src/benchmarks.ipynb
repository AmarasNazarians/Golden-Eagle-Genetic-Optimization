```python
D = 30
N = 40
pop_size = 30
upper_bound = zeros(D)
lower_bound = zeros(D)
max_it = 1000
all_funcs = opfunu.get_functions_based_classname("2017") 
model_name = ['GEGO','GEO',"PSO","GA"]
log_file="log_file3.txt"
res_file = "results3.txt"
freq = [5,10,50,100]
func = all_funcs[11](ndim = D)

name = func.name
dim = D
ub= func.ub
lb = func.lb
print(name,"\n")

with open(res_file, "a") as f:
    f.write(f"Function name {name}\n")
Y_his = zeros((4,N))
X_his = zeros((N,dim))    

for j in range(len(freq)):
    
    with open(res_file, "a") as f:
        f.write(f"frequancy: {freq[j]}\n")
        
    for i in range(N):
        model = GEGO(func=func.evaluate, n_dim=dim, size_pop=pop_size, max_iter=max_it, lb=lb, ub=ub,frequancy = freq[j])  
        best_x, best_y = model.run()
        Y_his[j][i] = best_y
    mean_Y = np.mean(Y_his[j])
    std_Y = np.std(Y_his[j])
    with open(res_file, "a") as f:
        f.write(f"mean_Y={mean_Y}, std_Y={std_Y} \n")
    print(f" mean_Y={mean_Y}, std_Y={std_Y} \n")
    
```

```python
N = 30
D = 30
max_it = 100
pop_size=20
all_funcs = unimodel_function
model_name = ["GEGO","GEO","PSO","GA","GWO","JSO","SCA"]
res_file = "results13.txt"
with open(res_file, "a") as f:
        f.write(f"max_it = {max_it}, pop_size:={pop_size}\n")

for name,func in all_funcs.items():   
    
    dim = func.ndim
    ub= func.ub
    lb = func.lb
    print(name,"\n")
    with open(res_file, "a") as f:
        f.write(f"Function name {name}, type: {func.unimodal}\n")
    Y_his = zeros(N)
    X_his = zeros((N,dim))    
    for j in range(len(model_name)):
        with open(res_file, "a") as f:
            f.write(f"model name: {model_name[j]}\n")
        for i in range(N):
            
            if(model_name[j] == "GEGO"):
                model = GEGO(func=func.evaluate, n_dim=dim, size_pop=pop_size, max_iter=max_it, lb=lb, ub=ub)
                best_x, best_y = model.run()
                Y_his[i] = best_y
            elif (model_name[j] == "GEO"):
                model = GEO(func=func.evaluate, n_dim=dim, pop=pop_size, max_iter=max_it, lb=lb, ub=ub)  
                best_x, best_y = model.run()
                Y_his[i] = best_y
            elif(model_name[j] == "PSO"):
                model = PSO(func=func.evaluate, n_dim=dim, pop=pop_size, max_iter=max_it, lb=lb, ub=ub)   
                best_x, best_y = model.run()
                Y_his[i] = best_y
            elif(model_name[j] == "GA"):
                model = EGA(func=func.evaluate, n_dim=dim, size_pop=pop_size, max_iter=max_it, lb=lb, ub=ub,n_elitist=4)    
                best_x, best_y = model.run() 
                Y_his[i] = best_y
            elif(model_name[j] == "GWO"):
                model = GWO(pack_size = pop_size, min_values = lb, max_values = ub, iterations = max_it, target_function = func.evaluate, verbose = False, start_init = None, target_value = None)
                minimum   = model[ -1]
                Y_his[i] = minimum 
            elif(model_name[j] == "JSO"):
                model = JSO(jellyfishes = pop_size, min_values = lb, max_values = ub, iterations = max_it, target_function = func.evaluate, verbose = False, start_init = None, target_value = None)
                minimum   = model[ -1]
                Y_his[i] = minimum  
            elif(model_name[j] == "SCA"):
                model = SCA(solutions = pop_size, min_values = lb, max_values = ub, iterations = max_it, target_function = func.evaluate, verbose = False, start_init = None, target_value = None)
                minimum   = model[ -1]
                Y_his[i] = minimum    
        
        mean_Y = np.mean(Y_his)
        std_Y = np.std(Y_his)
        with open(res_file, "a") as f:
            f.write(f"mean_Y={mean_Y}, std_Y={std_Y} \n")
        print(f" mean_Y={mean_Y}, std_Y={std_Y} \n")
    
    
```
